```python
import requests

# URL do arquivo README.md
url = "https://raw.githubusercontent.com/carlosantos2/curso_de_git_1/main/README.md"

# Baixar o conte√∫do do arquivo
response = requests.get(url)
if response.status_code == 200:
    # Salvar o arquivo localmente
    with open("README.md", "w") as file:
        file.write(response.text)
    print("Arquivo README.md baixado com sucesso!")
else:
    print("Falha ao baixar o arquivo. Verifique a URL.")
```

```python
from pyspark.sql import SparkSession

# Iniciar uma sess√£o Spark
spark = SparkSession.builder.appName("WordCount").getOrCreate()

# Ler o arquivo README.md
text_file = spark.read.text("README.md")

# Transformar os dados para contar as palavras
words = text_file.rdd.flatMap(lambda line: line.value.split(" "))
word_counts = words.map(lambda word: (word, 1)).reduceByKey(lambda a, b: a + b)

# Ordenar por frequ√™ncia (opcional)
sorted_word_counts = word_counts.sortBy(lambda x: x[1], ascending=False)

# Coletar todas as palavras e suas contagens
all_word_counts = sorted_word_counts.collect()

# Exibir todas as palavras e suas ocorr√™ncias
for word, count in all_word_counts:
    print(f"{word}: {count}")
```
